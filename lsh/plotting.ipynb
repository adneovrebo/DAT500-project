```python
import matplotlib.pyplot as plt
import numpy as np


'''
s: similarity
r: rows per band
b: number of bands
'''
def plotProbability(s, b, r):
   
   plt.figure(figsize=(15, 10))
   for i,b_i in enumerate(b):
      r_i = r[i]
      p = 1 - (1 - s**r_i)**b_i

      plt.plot(s, p, label=f'r: {r_i:.2f}, b: {b_i}', )

   # Add name on x-axis
   plt.xlabel('Similarity')
   # Add name on y-axis
   plt.ylabel('Probability')

   # Add legends with the names of the lines
   plt.legend(loc='best')
   # Change name on legend
   plt.title('Probability of finding a similar article')


s = np.arange(0.01, 1, 0.01)
b = np.array([100, 75, 50, 40, 25, 20, 10, 5, 4, 2, 1])
r = 100/b
plotProbability(s, b, r)
```
